```python
import xmlschema
import pprint
from requests import get

# Absent an XSD file for the new bdx-as4 conformance profile, 
# use the ebms-header-3_0-200704 XSD file

schema_xsd = get('https://docs.oasis-open.org/ebxml-msg/ebms/v3.0/core/ebms-header-3_0-200704.xsd')
schema_xsd = schema_xsd.content.decode('utf-8')

print_schema = 0
if print_schema == 1:
    print(f'Here\'s the schema from the Internet:')
    print(schema_xsd)

# Load the schema from the internet    
schema = xmlschema.XMLSchema('https://docs.oasis-open.org/ebxml-msg/ebms/v3.0/core/ebms-header-3_0-200704.xsd')
print(f'\n\nSchema w/namespace:\n{schema}\n\n\n')
print(f'Schema as loaded into XMLSchema is type: {type(schema)}')


# Print the entire list of schema components 
# See https://xmlschema.readthedocs.io/en/latest/components.html for an explanation of how the hierachy of components.
print_components = 1
if print_components == 1:
    print(f'\n\n\n\nComplete list of schema components...')
    for component in schema.iter_components():
        print(component)

        

print(f'\n\n\n\nComplete list of schema globals...')
for component in schema.iter_globals():
    print(component)    

      

print(f'\n\n\nSchema type: {type(schema)}')
print(f'Schema attributes:\n{dir(schema)}\n\n')
print(f'Schema root elements:\n{schema.root_elements}\n\n')
print(f'Schema validity:\n{schema.validity}\n\n')
    
    
schema_dict = {schema.name: schema.namespaces}   
print(schema_dict)

# Here is the schema defined and then applying it to a sample message.
schema.to_dict('./sample_msg.xml') 
```
