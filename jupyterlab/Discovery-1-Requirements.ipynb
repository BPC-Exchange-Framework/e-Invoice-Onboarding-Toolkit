## Assumptions ##

- All of our activities take place on a single access point.  
- Hosting an access point is a distinct service separate from the SML, which is a DNS lookup, the SMP, which uses a REST API, or the construciton of the AS4 envelope around the invoice to forward to the second access point.
- XML logging using agreed upon guid or other tracking ID.  This will help us track the message through the acccess point.  
- A completed record of a trip for an invoice will have 10 messages (4 from SML component, 4 from SNP component, 2 from delivery componemt)
- Other logging to be done at the applicaton level.  
- Log handling will need to be offloaded to either a database or file system on the server, or on the cloud.  No matter what, any entity that participates in this process by offering an access point is going to assume some infrastructure overhead.
